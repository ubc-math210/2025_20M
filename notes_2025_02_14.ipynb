## Vectorization

Vectorization means applying an operation or a function to each entry of a vector elementwise. For example, if $\mathbf{x}$ is a vector:

$$
\mathbf{x} = \begin{bmatrix} x_0 & x_1 & x_2 & \cdots & x_N \end{bmatrix}
$$

and $f(x)$ is a function then the vectorized function applies $f$ to each entry of $\mathbf{x}$ elementwise:

$$
f(\mathbf{x}) = \begin{bmatrix} f(x_0) & f(x_1) & f(x_2) & \cdots & f(x_N) \end{bmatrix}
$$

We have already seen vectorized operations in linear algebra like addition and scalar multiplication:

$$
\mathbf{u} + \mathbf{v} =  \begin{bmatrix} u_0 + v_0 & u_1 + v_1 & u_2 + v_2 & \cdots & u_N + v_N \end{bmatrix}
$$

$$
c\mathbf{u} =  \begin{bmatrix} cu_0 & cu_1 & cu_2 & \cdots & cu_N \end{bmatrix}
$$

Vectorized arithmetic operations are

$$
\mathbf{x}^2 = \begin{bmatrix} x_0^2 & x_1^2 & x_2^2 & \cdots & x_N^2 \end{bmatrix}
$$

$$
\mathbf{x} + c = \begin{bmatrix} x_0 + c & x_1 + c & x_2 + c & \cdots & x_N + c \end{bmatrix}
$$

$$
\mathbf{x} \mathbf{y} = \begin{bmatrix} x_0y_0 & x_1y_1 & x_2y_2 & \cdots & x_Ny_N \end{bmatrix}
$$

## Plotting Procedure

1. Generate vector of $x$ values
2. Use vector functions and operations to construct vector of $y$ values.
3. `plt.plot(x,y)` then add style, title, labels, etc.
4. `plt.show()`

Use NumPy function `np.linspace(a,b,N)` to create the vector of $N$ evenly spaced values from `a` to `b` (inclusive). Note that the space between values is

$$
\frac{b - a}{N - 1}
$$

```python
x = np.linspace(0,2*np.pi,201)
y = np.sin(x)
plt.plot(x,y)
plt.title('Sine function'), plt.xlabel('x values'), plt.ylabel('y values'), plt.grid(True)
plt.show()
```

## Examples: Solutions of ODEs

Plot the unique solution of $y' = -y^2 t$ for each initial value $y(0) = 0,0.1,0.2,...,2.0$.

```python
t = np.linspace(0,5,200)

for y0 in np.arange(0.1,2.1,0.1):
    y = 1/(t**2/2 + 1/y0)
    plt.plot(t,y,'b')
    
plt.plot(t,0*t,'b')
plt.grid(True)
plt.show()
```

Plot the unique solution of $y'' + 2y' + 2y = 0, y(0) = 0$ for each initial value $y'(0) = 0,0.1,0.2,...,2.0$.

```python
a = 1
b = 2
c = 2
D = b**2 - 4*a*b
print(D)
```

```python
alpha = -b/(2*a)
beta = abs(D)**0.5/(2*a)
```

```python
t = np.linspace(0,5,200)
for dy0 in np.arange(0,2.1,0.1):
    y = dy0/beta*np.exp(alpha*t)*np.sin(beta*t)
    plt.plot(t,y,'b')
plt.grid(True)
plt.show()
```
