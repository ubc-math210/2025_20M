## Riemann Sums

Compute left and right Riemann sums to approximate:

$$
\int_0^1 e^{-x^2} dx
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2,2,500)
y = np.exp(-x**2)
plt.figure(figsize=(10,3))
plt.plot(x,y), plt.grid(True)
plt.show()
```

```python
a = 0
b = 1
N = 10
x = np.linspace(a,b,N+1)
y = np.exp(-x**2)
dx = (b - a)/N
R = np.sum(y[1:]*dx)
print(R)
```

```python
L = np.sum(y[:-1]*dx)
print(L)
```

```python
(L + R)/2
```
