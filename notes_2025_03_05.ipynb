## Numerical Integration

Consider the logarithmic integral

$$
\int_e^{e^2} \frac{dx}{\ln(x)}
$$

Find $N$ which gurantees the error is less than $0.001$ for: (a) left Riemann sum and (b) trapezoid rule.

```python
a = np.exp(1)
b = np.exp(2)
f = lambda x: 1/np.log(x)
df = lambda x: -1/x/np.log(x)**2
d2f = lambda x: (np.log(x) + 2)/x**2/np.log(x)**3
```

```python
x = np.linspace(a,b,200)
plt.plot(x,f(x)), plt.ylim([0,1.5])
plt.show()
```

```python
x = np.linspace(a,b,200)
plt.plot(x,df(x))
plt.show()
```

```python
x = np.linspace(a,b,200)
plt.plot(x,d2f(x))
plt.show()
```

```python
N = 4013
x = np.linspace(a,b,N+1)
y = f(x)
dx = (b - a)/N
LN = dx*np.sum(y[:-1])
print(LN)
```

```python
N = 59
x = np.linspace(a,b,N+1)
y = f(x)
dx = (b - a)/N
TN = dx*np.sum(y[:-1] + y[1:])/2
print(TN)
```

```python
np.trapz(y,x)
```

```python
TN + 0.001
```

```python
TN - 0.001
```
