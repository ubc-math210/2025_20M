## Order of Accuracy

```python
def odeRK3(f,t,y0):
    N = len(t) - 1
    y = np.zeros(N + 1)
    y[0] = y0
    for n in range(N):
        h = t[n + 1] - t[n]
        k1 = f(t[n],y[n])
        k2 = f(t[n] + h/3,y[n] + k1*h/3)
        k3 = f(t[n] + 2*h/3, y[n] + k2*2*h/3)
        y[n + 1] = y[n] + (k1/4 + 3*k3/4)*h
    return y
```

```python
f = lambda t,y: y
y0 = 1
t = np.linspace(0,1,21)
y = odeRK3(f,t,y0)
plt.plot(t,y,'.-')

ts = np.linspace(0,1,200)
ys = np.exp(ts)
plt.plot(ts,ys)
plt.legend(['RK3','Exact'])
plt.show()
```

```python
hs = [0.1,0.05,0.01,0.005,0.001]
Es = []
for k in range(len(hs)):
    y = odeRK3(f,[0,hs[k]],y0)
    E = np.abs(np.exp(hs[k]) - y[1])
    Es.append(E)
plt.loglog(hs,Es,'.-'), plt.grid(True)
plt.show()
```
